```python
import skimage
from skimage import io

import matplotlib.pyplot as plt
import numpy as np



```

```python
def color2grey(img):
    b = [.3, .6, .1]
    return np.dot(img[...,:3], b)
```

```python
img = io.imread('houndog1.png')

print(img.shape)
plt.imshow(img)
plt.title('Dog')
plt.show()

gray_img = color2grey(img)
print(gray_img.shape)
plt.imshow(gray_img, cmap='gray')
plt.show()
```
